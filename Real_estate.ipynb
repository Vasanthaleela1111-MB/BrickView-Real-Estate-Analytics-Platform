Connecting to Sqlite3

```python
import sqlite3
conn=sqlite3.connect("brickview_realestate.db")
cursor=conn.cursor()
```

```python
create_listings="""
CREATE TABLE IF NOT EXISTS listings(
    listing_id INTEGER PRIMARY KEY,
    address TEXT,
    city TEXT,
    zip TEXT,
    property_type TEXT,
    price INT,
    area_sqft REAL,
    agent_id INTEGER,
    listed_date DATE
);
"""
cursor.execute(create_listings)
conn.commit()
```

```python
cursor.execute("PRAGMA table_info(listings);")
columns=cursor.fetchall()
for col in columns:
    print(col)
```

```python
create_property_attributes="""
CREATE TABLE IF NOT EXISTS property_attributes(
    attribute_id INTEGER PRIMARY KEY,
    listing_id INT,
    bedrooms INT,
    bathrooms INT,
    floor_number INT,
    total_floors INT,
    year_built DATE,
    is_rented BOOL,
    tenant_count INT,
    furnishing_status TEXT,
    metro_distance_km REAL,
    parking_available BOOL,
    power_backup BOOL,
    FOREIGN KEY (listing_id) REFERENCES listings(listing_id)
);
"""
cursor.execute(create_property_attributes)
conn.commit()
```

```python
create_table_agents="""
CREATE TABLE IF NOT EXISTS agents(
    Agent_ID TEXT PRIMARY KEY,
    Name TEXT,
    City TEXT,
    Contact TEXT,
    Commission_Rate REAL,
    Deals_Closed INT,
    Rating REAL,
    Experience_Years INT,
    Avg_Closing_Days INT
);
"""
cursor.execute(create_table_agents)
conn.commit()
```

```python
create_table_sales="""
CREATE TABLE IF NOT EXISTS sales(
    Sale_ID TEXT PRIMARY KEY,
    listing_id INT,
    Sale_Date DATE,
    Sale_Price INT,
    Days_On_Market INT,
    FOREIGN key (listing_id) REFERENCES sales(listing_id)
);
"""
cursor.execute(create_table_sales)
conn.commit()
```

```python
create_table_buyers="""
CREATE TABLE IF NOT EXISTS buyers(
    Buyer_ID INT PRIMARY KEY,
    Sale_ID INT,
    Buyer_Type TEXT,
    Payment_Mode TEXt,
    Loan_Taken BOOL,
    Loan_Provider TEXT,
    Loan_Amount INT
);
"""
cursor.execute(create_table_buyers)
conn.commit()
```

```python
list.to_sql("listings",conn,if_exists="replace",index=False)
prop.to_sql("property_attributes",conn,if_exists="replace",index=False)
agents_20.to_sql("agents",conn,if_exists="replace",index=False)
agents_enhanced.to_sql("agents_enhanced",conn,if_exists="replace",index=False)
sales.to_sql("sales",conn,if_exists="replace",index=False)
buy.to_sql("buyers",conn,if_exists="replace",index=False)
```

```python
query="""
select p.furnishing_status,AVG(l.price) as avg_price from listings l INNER JOIN property_attributes p ON l.listing_id = p.listing_id group by p.furnishing_status order by avg_price;
"""
furnish=pd.read_sql(query,conn)
print("How does furnishing status impact property prices?")
print(furnish)
```

```python
query="""
select ROUND(p.metro_distance_km) AS metro_distance,AVG(l.price) AS price from listings l INNER JOIN property_attributes p ON l.listing_id=p.listing_id  group by ROUND(p.metro_distance_km) order by metro_distance;
"""
a=pd.read_sql(query,conn)
print("Do properties closer to metro stations command higher prices?")
print(a)
```

```python
query="""
select p.is_rented,AVG(l.price)as price from listings l INNER JOIN property_attributes p ON l.listing_id=p.listing_id group by p.is_rented
"""
rented=pd.read_sql(query,conn)
print("Are rented properties priced differently from non-rented ones?")
print(rented )
```

```python
query="""
select p.bedrooms,p.bathrooms,avg(price) as price from listings l INNER JOIN property_attributes p on l.listing_id=p.listing_id group by p.bedrooms,p.bathrooms order by p.bedrooms,p.bathrooms
"""
b=pd.read_sql(query,conn)
print("How do bedrooms and bathrooms affect pricing?")
print(b)
```

 Sales & Market Performance


```python
query="""
select l.city,avg(Days_on_Market) as days from sales s inner join listings l on s.Listing_ID=l.Listing_ID group by l.city
"""
d=pd.read_sql(query,conn)
print("What is the average days on market by city?")
print(d)
```

```python
query="""
select l.property_type,round(avg(s.Days_on_Market)) as days from sales s inner join listings l on s.Listing_ID	= l.Listing_ID group by l.property_type order by days
"""
d=pd.read_sql(query,conn)
print("Which property types sell the fastest?")
print(d)
```

```python
query="""
SELECT 
    (COUNT(*) * 100.0 / 
        (SELECT COUNT(*) 
         FROM sales s
         JOIN listings l 
           ON s.listing_id = l.listing_id)
    ) AS percentage_higher_sales
FROM sales s
JOIN listings l 
  ON s.listing_id = l.listing_id
WHERE s.sale_price > l.price;"""
d=pd.read_sql(query,conn)
print("What percentage of properties are sold above listing price?")
print(d)
```

```python
query="""
select avg(s.Sale_Price)/(l.price) l.city from sales s inner join listings l on s.Sale_Price=l.price
"""
d=pd.read_sql(query,conn)
print("What is the sale-to-list price ratio by city?")
print(d)
```

```python
query="""
select l.Listing_ID,l.City,l.Property_Type,l.Price,s.Sale_Price,s.Days_on_Market from sales s inner join listings l on s.Listing_ID=l.Listing_ID where s.Days_on_Market>90 order by s.Days_on_Market
"""
d=pd.read_sql(query,conn)
print("Which listings took more than 90 days to sell?")
d
```

```python
query="""
select round(avg(Days_on_Market)) as days
,round(metro_distance_km) as metro_distance from sales s inner join property_attributes p on s.listing_id = p.listing_id group by metro_distance
"""
d=pd.read_sql(query,conn)
print("How does metro distance affect time on market?")
d
```

```python
query="""
select Date_Sold,Sale_Price avg(price) from sales group by Date_Sold
"""
d=pd.read_sql(query,conn)
print("What is the monthly sales trend?")
print(d)
```

```python
query="""
select * from listings where Listing_ID not null
"""
d=pd.read_sql(query,conn)
print("Which properties are currently unsold?")
print(d)
```

 Buyer & Financing Behavior


```python
query="""
select buyer_type,count(buyer_type)/100 from buyers GROUP BY buyer_type
"""
d=pd.read_sql(query,conn)
print("What percentage of buyers are investors vs end users?")
print(d)
```

```python
query="""
SELECT 
    l.City,AVG(b.loan_taken ) AS loan_rate FROM buyers b JOIN listings l ON b.sale_id = l.Listing_ID GROUP BY l.City order by loan_rate"""
d=pd.read_sql(query,conn)
print("Which cities have the highest loan uptake rate?")
print(d)
```

```python
query="""
select buyer_type,round(avg(loan_amount)) as amount from buyers WHERE loan_taken = 1 group by buyer_type
"""
g=pd.read_sql(query,conn)
print("What is the average loan amount by buyer type?")
g
```

```python
query="""
select payment_mode,count(*) as commonly_used from buyers group by payment_mode
"""
g=pd.read_sql(query,conn)
print("Which payment mode is most commonly used?")
g
```
